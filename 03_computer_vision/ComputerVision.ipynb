Use Numpy's .reshape() method to covert the training and testing data from a matrix into an vector of features. So, X_test[index].reshape((1,784)) will convert the index -th element of X_test into a 28x28 = 784 -length row vector of values, rather than a matrix.
```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

N_train, N_test = X_train.shape[0], X_test.shape[0]
X_train_reshaped = X_train.reshape(N_train, 784) 
X_test_reshaped = X_test.reshape(N_test, 784)     

print(f"X_train_reshaped shape: {X_train_reshaped.shape}")
print(f"X_test_reshaped shape: {X_test_reshaped.shape}")
```

Use the reshaped X_test and y_test data to create a k-nearest neighbor classifier of digit. What is the optimal number of neighbors k? If you can't determine this, play around with different values of k for your classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
k_values = range(1, 20) 
accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_reshaped, y_train)
    y_pred = knn.predict(X_test_reshaped)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)
    print(f"k={k}, Accuracy={accuracy:.4f}")

# Plot accuracy vs. k
plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker='o')
plt.title("Accuracy vs. Number of Neighbors (k)")
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.grid()
plt.show()
```
